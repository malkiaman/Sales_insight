# 📊 Sales Insights Dashboard - README

Welcome to the **Sales Insights Dashboard** project, originally created in Power BI and now documented for replication or extension using Python.

---

## 🧾 Overview

This project focuses on analyzing sales data to derive actionable insights. The dashboard summarizes key metrics like total revenue and sales quantity, highlights top-performing markets, products, and customers, and shows revenue trends over time.

---

## 📌 Key Features

- **Total Revenue and Sales Quantity KPIs**
- **Revenue by Market**
- **Sales Quantity by Market**
- **Revenue Trend Over Time**
- **Top 5 Products**
- **Top 5 Customers**

---

## 🗂 Dataset Structure

- `Date`: Transaction date
- `Market`: Market name or region
- `Revenue`: Sales revenue
- `Sales_Qty`: Quantity sold
- `Product`: Product identifier or name
- `Customer`: Customer name
